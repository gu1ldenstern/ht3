```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.gridspec as gridspec
from scipy.interpolate import interp1d
from matplotlib.ticker import MultipleLocator



# придумываю как нормально и красиво нарисовать кривую, если про нее ничего неизвестно
# лучшее, что придумалось - сначала наметить ее по точкам, а после интерполировать
x = np.linspace(-100, 1000, num=21, endpoint=True)
y = [0,0,0,0,0.1,1,3,5,2,5,7,3,4,3,3,3,3,3,2,4,5]
f2 = interp1d(x, y, kind='quadratic')
xnew = np.linspace(-100, 1000, num=1100, endpoint=True)


# сездаем сабплоты
fig, axs = plt.subplots(2, figsize=(12, 10))


#ПЕРВЫЙ САБПЛОТ

# рисуем первый график - который придумали с помощью интерполяции
axs[0].plot(xnew, f2(xnew), color = 'black')

# читерим на втором графике - рисуем его обратным первому. ну в принципе похож
axs[0].plot(xnew, 1/(f2(xnew)+0.5), color = 'red')

# рисуем прямые параллельные оси OY
axs[0].axvline(x=0, color = 'black', linestyle = 'dotted')
axs[0].axvline(x=76, color = 'black')
axs[0].axvline(x=200, color = 'black')
axs[0].axvline(x=276, color = 'black')
axs[0].axvline(x=416, color = 'black')
axs[0].axvline(x=508, color = 'black')
axs[0].axvline(x=640, color = 'black')

# пишем римские цифры посреди промежутков образованных параллельными осями
axs[0].text(276/2, 6.5,'I.', fontsize = 14)
axs[0].text(200+76/2, 6.5,'II.', fontsize = 14)
axs[0].text(276 + (416-276)/2, 6.5,'III.', fontsize = 14)
axs[0].text(416 + (508-416)/2, 6.5,'IV.', fontsize = 14)
axs[0].text(508 + (640-508)/2, 6.5,'V.', fontsize = 14)

# пишем подписи
axs[0].text(-100, 7.5, 'GFP', color = 'black', fontweight = 'bold', fontsize = 14)
axs[0].text(-30, 7.5, 'Diss*10', color = 'red', fontweight = 'bold', fontsize = 14)

axs[0].text(980, 0.3, '[ms]', color = 'black')
axs[0].text(-90, 7.1, '[μV]', color = 'black')

# пишем название
# это должен быть не таймс нью роман а какой-то другой шрифт без засечек, но суть понятна
hfont = {'fontname':'Times New Roman'}
axs[0].set_title('Correct Go', fontsize =22, **hfont)

# ставим засечки на осях с определенным промежутком и определяем промежуток для написания цифр на оси OX
axs[0].xaxis.set_major_locator(MultipleLocator(100))
axs[0].xaxis.set_minor_locator(MultipleLocator(5))
axs[0].yaxis.set_minor_locator(MultipleLocator(0.1))

#заносим цифры внутрь графика
axs[0].tick_params(axis="y",direction="in", pad=-22)
axs[0].tick_params(axis="x",direction="in", pad=-15)

#определяем диапозон в котором будут ставиться ticks
axs[0].set_xlim(-99, 999)
axs[0].set_ylim(0, 7.9)

#делаем невидимыми границы сверху и справа
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)


#ВТОРОЙ САБПЛОТ

# рисуем сабплот такой же как первый. можно в принципе и другой нарисовать но мне и так норм, а вам?
axs[1].plot(xnew, f2(xnew), color = 'black')
axs[1].plot(xnew, 1/(f2(xnew)+0.5), color = 'red')

# рисуем прямые параллельные оси OY
axs[1].axvline(x=0, color = 'black', linestyle = 'dotted')
axs[1].axvline(x=76, color = 'black')
axs[1].axvline(x=200, color = 'black')
axs[1].axvline(x=276, color = 'black')
axs[1].axvline(x=432, color = 'black')
axs[1].axvline(x=508, color = 'black')
axs[1].axvline(x=596, color = 'black')
axs[1].axvline(x=724, color = 'black')

# пишем римские цифры посреди промежутков образованных параллельными осями
axs[1].text(276/2, 6.5,'I.', fontsize = 14)
axs[1].text(200+76/2, 6.5,'II.', fontsize = 14)
axs[1].text(276 + (432-276)/2, 6.5,'III.', fontsize = 14)
axs[1].text(432 + (508-432)/2, 6.5,'IV.', fontsize = 14)
axs[1].text(508 + (596-508)/2, 6.5,'V.', fontsize = 14)
axs[1].text(596 + (724-596)/2, 6.5,'VI.', fontsize = 14)

# пишем подписи
axs[1].text(-100, 7.5, 'GFP', color = 'black', fontweight = 'bold', fontsize = 14)
axs[1].text(-30, 7.5, 'Diss*10', color = 'red', fontweight = 'bold', fontsize = 14)

axs[1].text(980, 0.3, '[ms]', color = 'black')
axs[1].text(-90, 7.1, '[μV]', color = 'black')

# пишем название
axs[1].set_title('Successful Stop', fontsize =22, **hfont)

# ставим засечки на осях с определенным промежутком и определяем промежуток для написания цифр на оси OX
axs[1].xaxis.set_major_locator(MultipleLocator(100))
axs[1].xaxis.set_minor_locator(MultipleLocator(5))
axs[1].yaxis.set_minor_locator(MultipleLocator(0.1))

#заносим цифры внутрь графика
axs[1].tick_params(axis="y",direction="in", pad=-22)
axs[1].tick_params(axis="x",direction="in", pad=-15)

#определяем диапозон в котором будут ставиться ticks
axs[1].set_xlim(-99, 999)
axs[1].set_ylim(0, 7.9)

#делаем невидимыми границы сверху и справа
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)

plt.show()
```
